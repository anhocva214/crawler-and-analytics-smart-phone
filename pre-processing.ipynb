```python
df.columns
```

```python
new_columns = ['brand', 'name', 'ratting', 'ratting_total', 'price_sale',
       'price_origin', 'status', 'Model', 'Màu sắc', 'Nhà sản xuất',
       'Xuất xứ', 'Thời gian bảo hành', 'Địa điểm bảo hành', 'Chipset',
       'RAM', 'Bộ nhớ trong', 'Hệ điều hành']
df.columns = new_columns
df.columns
```

```python
# Sum missing value each column
for col in df.columns:
    print(col, "=",  df[col].isna().sum())
```

```python
for col in df.columns:
    df[col] = df[col].apply(lambda x:  x.strip() if type(x) == type('str') else x)
    
df.price_origin = df.price_origin.apply(lambda x: int(x.replace('đ', '').replace('.','').strip()))
df.price_sale = df.price_sale.apply(lambda x: int(x.replace('đ', '').replace('.','').strip()))

# df.head()
```

#### Bộ nhớ trong

```python
df['Bộ nhớ trong'].unique()
```

```python
storage = df['Bộ nhớ trong'].values
# storage
for index, value in enumerate(storage):
    # print(value, " - ", type(value))
    if type(value)==type('str'):
        x = value.replace('GB','')
        arr = x.split(" ")
        result = 0
        if len(arr) == 1:
            result = int(arr[0])
        else:
            for text in arr:
                if text.isnumeric() == True:
                    result = int(text)
                    break
        storage[index] = result
    else:
        storage[index] = 0
        
df['Bộ nhớ trong'] = storage
df['Bộ nhớ trong'].unique()
```

```python
storage = df['RAM'].values
# storage
for index, value in enumerate(storage):
    # print(value, " - ", type(value))
    if type(value)==type('str'):
        x = value.replace('GB','')
        arr = x.split(" ")
        result = 0
        if len(arr) == 1:
            result = int(arr[0])
        else:
            for text in arr:
                if text.isnumeric() == True:
                    result = int(text)
                    break
        storage[index] = result
    else:
        storage[index] = 0
        
df['RAM'] = storage
df['RAM'].unique()
```

#### Hệ điều hành

```python
df['Hệ điều hành'].unique()
```

```python
const_ios = ['android', 'ios']
values_ios = df['Hệ điều hành'].values
for index, value in enumerate(values_ios):
    if type(value) == type('str'):
        if value.lower().find(const_ios[0]) >=0 :
            values_ios[index] = const_ios[0]
        elif value.lower().find(const_ios[1]) >=0 :
            values_ios[index] = const_ios[1]
        elif value.lower().find('Funtouch OS'.lower()) >=0:
            values_ios[index] = const_ios[0]
    
df['Hệ điều hành'] = values_ios
```

```python
df[df['Hệ điều hành'].isna() == True]
```

=> Những sản phẩm trên không phải **iphone** nên hệ điều thành sẽ là **android**

```python
df['Hệ điều hành'] = df['Hệ điều hành'].fillna('android')
```

#### ratting_total

```python
df['ratting_total'].unique()
```

```python
ratting_values = df['ratting_total'].values
for index, value in enumerate(ratting_values):
    arr = value.split(" ")
    for text in arr:
        if text.isnumeric() == True:
            ratting_values[index] = int(text)
            
df['ratting_total'] = ratting_values
```

#### brand

```python
df['brand'].unique()
```

```python
df['brand'] = df['brand'].apply(lambda x: x.replace('Điện thoại', '').strip())
```

#### name

```python
name_values = df['name'].values
for index, value in enumerate(name_values):
    name = value.replace("Điện thoại", "")
    # brand = df.iloc[index]['brand'].lower()
    color = df.iloc[index]['Màu sắc']
    name = name.replace(color, "").replace("/", "")
    for value in [128, 256, 64, 32, 512]:
        name = name.replace(str(value)+"GB", "")
    for value in [8, 6, 4, 12, 3, 2]:
        name = name.replace(str(value)+"GB", "")
    name_values[index] = name
    # print(name)

df['name'] = name_values  
```
