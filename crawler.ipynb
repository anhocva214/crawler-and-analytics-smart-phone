```python
import json

class Phone:
    name = ''
    brand = ''
    ratting = ''
    price = ''
    price_origin = ''
    insurance = ''
    screen_tech= ''
    resolution = ''
    screen_size = ''
    quality_video = ''
    flash_light= ''
    os = ''
    chipset = ''
    cpu_speed = ''
    gpu = ''
    ram = ''
    storage = ''
    network = ''
    sim = ''
    bluetooth = ''
    charge = ''
    jack_headphone = ''
    design=''
    size=''
    weight=''
    battery=''
    battery_type=''
    battery_tech = ''

    def __iter__(self):
        # first start by grabbing the Class items
        iters = dict((x,y) for x,y in Phone.__dict__.items() if x[:2] != '__')

        # then update the class items with the instance items
        iters.update(self.__dict__)

        # now 'yield' through the items
        for x,y in iters.items():
            yield x,y
```
