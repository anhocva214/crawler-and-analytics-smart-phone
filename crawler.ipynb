```python
def getInfo(product_link):
    
    page_item = requests.get(product_link)
    soup_item = BeautifulSoup(page_item.content, "html.parser")
    phone = {}

    def getTextSelect(str):
        return soup_item.select_one(str).get_text()

    phone['brand'] = getTextSelect('#tygh_main_container > div.tygh-content.clearfix > div > div:nth-child(1) > div > div.container > div > div > div.NkPdp_productInfo > div.productInfo_col-23 > div.productInfo_col-23-top > div.pdp_breadcrumbs > a.last-item')
    phone['name'] = getTextSelect('#tygh_main_container > div.tygh-content.clearfix > div > div:nth-child(1) > div > div.container > div > div > div.NkPdp_productInfo > div.productInfo_col-23 > div.productInfo_col-23-top > div:nth-child(2) > h1')
    try:
        phone['ratting'] = getTextSelect('#average_rating_product > span.number_avg_rate_npv')
    except:
        phone['ratting'] = '0'
        
    try:
        phone['ratting_total'] = getTextSelect('#tygh_main_container > div.tygh-content.clearfix > div > div:nth-child(1) > div > div.container > div > div > div.NkPdp_productInfo > div.productInfo_col-23 > div.productInfo_col-23-top > div.wrap_fb.wrap_name_vote > div > a')
    except:
        phone['ratting_total'] = '0'
    
    phone['price_sale'] = getTextSelect('#tygh_main_container > div.tygh-content.clearfix > div > div:nth-child(1) > div > div.container > div > div > div.NkPdp_productInfo > div.productInfo_col-23 > div.productInfo_col-2 > form > div.nk2020-pdp-price > div.l > div.product_info_price > div > div > div.product_info_price_value-final > span')
    
    try:
        phone['price_origin'] = getTextSelect('#tygh_main_container > div.tygh-content.clearfix > div > div:nth-child(1) > div > div.container > div > div > div.NkPdp_productInfo > div.productInfo_col-23 > div.productInfo_col-2 > form > div.nk2020-pdp-price > div.l > div.product_info_price > div > div > div.price_promotion > div > span')
    except:
        phone['price_origin'] = phone['price_sale']
    
    phone['status'] = getTextSelect('#tygh_main_container > div.tygh-content.clearfix > div > div:nth-child(1) > div > div.container > div > div > div.NkPdp_productInfo > div.productInfo_col-23 > div.productInfo_col-2 > form > div.nk2020-pdp-price > div.l > div.nk-top-stickers > span.nk-sticker.nk-new')

    features = soup_item.select('#content_features > div.NkPdp_productSpecification > div.productSpecification_col-2 > div > table > tbody > tr')
    for f in features:
        name = f.select_one('td.title').get_text()
        value = f.select_one('td.value').get_text()
        phone[name] = value
        

    return dict(phone)


# getInfo('https://didongviet.vn/oppo-a16k.html')
```

```python
data_dict = []

for page in range(1, 6):
    URL = "https://www.nguyenkim.com/dien-thoai-di-dong/page-{}".format(page)
    page = requests.get(URL)

    soup = BeautifulSoup(page.content, "html.parser")
    
    product_list = soup.select('#pagination_contents .item.nk-fgp-items.nk-new-layout-product-grid')
    # print(product_list[0])

    for product in product_list:
        product_link =  product.find('a').get_attribute_list('href')[0]
        print(product_link)
        data_dict.append(getInfo(product_link))

        
        
```

```python
df = pd.DataFrame.from_dict(data_dict, orient='columns')
df.to_csv('data_phone_nguyen_kiem.csv')
```
