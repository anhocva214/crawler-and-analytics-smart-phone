```python
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('./data_phone_1_to_55.csv')
df2 = pd.read_csv('./data_phone_apple_1_to_56.csv')
df = pd.concat([df1, df2])
# df = df.drop_duplicates(subset=['name'], keep=False)
df.info()
```

```python
cols_new = []
for col in df:
    # print(col)
    if sum(pd.isnull(df[col])) < 150:
        cols_new.append(col)

cols_new
```

```python
df_new = df[cols_new].copy()
df_new = df_new.drop(['ratting', 'Quay phim', 'Unnamed: 0'], axis=1)
df_new.head()
```

## Price

```python
df_new['price'] = df_new['price'].fillna(0)
sum(df_new['price'].isnull())
```

```python
def handlePrice(value):
    value = str(value)
    value = value.replace('\xa0₫', '')
    value = value.replace('\n', '')
    value = value.replace('.', '')
    value = value.replace(' ', '')
    return float(value)
    

df_new['price'] = df_new['price'].apply(lambda x: handlePrice(x))
df_new['price']
```

```python
df_new[df_new['price'] == 0]
```

```python
brands =  df_new[df_new['price'] == 0]['brand'].unique()
brands
```

```python
for b in brands:
    price_mean_iphone =  df_new[df_new['brand'] == b]['price'].mean()
    print(price_mean_iphone)
    df_new.loc[(df_new['price'] == 0) & (df_new['brand'] == b), 'price'] = price_mean_iphone
```

```python
df_new[df_new['brand'] == 'Bphone']
```

```python
df_new.drop(df_new[df_new['brand'] == 'Bphone'].index, inplace = True)
```
